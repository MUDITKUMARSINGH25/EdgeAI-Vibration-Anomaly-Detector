```python
# @title Edge AI Vibration Anomaly Detection
# @markdown # Real-Time Vibration Anomaly Detection Simulator
# @markdown This project simulates an Arduino-based system for detecting anomalies in vibration sensor data, implementing concepts from TinyML and edge computing.

# Install required packages
!pip install numpy matplotlib

import numpy as np
import matplotlib.pyplot as plt
import json
import time
from IPython.display import clear_output

# @title Data Simulation Functions

def generate_vibration_data(length=1000, frequency=5.0, noise_level=0.2,
                           include_anomaly=True, anomaly_start=700, anomaly_length=100):
    """
    Generates a realistic vibration sensor signal.
    """
    time_vec = np.linspace(0, 10, length)
    # Base signal: a simple sine wave representing normal vibration
    base_signal = np.sin(2 * np.pi * frequency * time_vec)

    # Add some random noise (always present in real sensors)
    noise = noise_level * np.random.randn(length)
    signal = base_signal + noise

    # Simulate a potential fault (a sudden change in amplitude and frequency)
    if include_anomaly:
        anomaly_slice = slice(anomaly_start, anomaly_start + anomaly_length)
        signal[anomaly_slice] = 2.0 * np.sin(2 * np.pi * 15.0 * time_vec[anomaly_slice]) + 0.5 * np.random.randn(anomaly_length)

    return signal

# @title Anomaly Detection Class

class AnomalyDetector:
    def __init__(self, window_size=10, threshold=2.5):
        self.window_size = window_size
        self.threshold = threshold  # k value, higher = less sensitive
        self.baseline_mean = None
        self.baseline_std = None

    def train(self, data):
        """Establishes the baseline 'normal' parameters from training data."""
        print("Training anomaly detector...")
        self.baseline_mean = np.mean(data)
        self.baseline_std = np.std(data)
        print(f"Baseline established: Mean = {self.baseline_mean:.4f}, Std = {self.baseline_std:.4f}")
        return self

    def check_anomaly(self, data_window):
        """Checks a window of data for anomalies against the trained baseline."""
        if self.baseline_mean is None:
            raise ValueError("Detector must be trained before use.")

        current_mean = np.mean(data_window)
        current_std = np.std(data_window)

        # Check if current stats deviate from baseline beyond the threshold
        mean_diff = abs(current_mean - self.baseline_mean)
        std_diff = abs(current_std - self.baseline_std)

        if mean_diff > self.threshold * self.baseline_std or std_diff > self.threshold * self.baseline_std:
            return True, current_mean, current_std
        else:
            return False, current_mean, current_std

# @title Generate Training Data
print("Generating training data (normal operation)...")
training_data = generate_vibration_data(include_anomaly=False)

# @title Train the Anomaly Detection Model
detector = AnomalyDetector()
detector.train(training_data)

# @title Generate Test Data with Anomaly
print("\nGenerating test data with anomaly...")
test_data = generate_vibration_data(include_anomaly=True)

# @title Real-time Simulation
print("Starting real-time simulation...")
print("This will simulate reading from a sensor and detecting anomalies")
print("Press the stop button to interrupt the simulation")

# Initialize variables for the simulation
window_size = 10
anomaly_flags = []
means = []
stds = []

try:
    for i in range(len(test_data) - window_size):
        # Get the current window of data
        window = test_data[i:i+window_size]

        # Check for anomalies
        is_anomaly, current_mean, current_std = detector.check_anomaly(window)
        anomaly_flags.append(is_anomaly)
        means.append(current_mean)
        stds.append(current_std)

        # Display status
        if is_anomaly:
            status = "ðŸš¨ ANOMALY DETECTED!"
        else:
            status = "âœ… Normal"

        clear_output(wait=True)
        print(f"Sample {i}: {status}")
        print(f"Current value: {test_data[i]:.4f}")
        print(f"Window mean: {current_mean:.4f}, std: {current_std:.4f}")
        print("\nPress the stop button to end simulation early")

        # Add a small delay to simulate real-time processing
        time.sleep(0.1)

except KeyboardInterrupt:
    print("Simulation stopped by user")

print("Simulation completed!")

# @title Visualization of Results

# Pad the lists to match the original data length for plotting
anomaly_flags_padded = [False] * (window_size // 2) + anomaly_flags + [False] * (window_size // 2)
means_padded = [np.mean(test_data[:window_size])] * (window_size // 2) + means + [means[-1]] * (window_size // 2)
stds_padded = [np.std(test_data[:window_size])] * (window_size // 2) + stds + [stds[-1]] * (window_size // 2)

# Create the plot
plt.figure(figsize=(12, 8))

# Plot the raw data
plt.subplot(3, 1, 1)
plt.plot(test_data, 'b-', label='Vibration Signal', alpha=0.7)
plt.title('Simulated Vibration Sensor Data with Anomaly')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)

# Plot the moving statistics
plt.subplot(3, 1, 2)
plt.plot(means_padded, 'g-', label='Moving Mean', linewidth=2)
plt.axhline(y=detector.baseline_mean, color='k', linestyle='--', label='Baseline Mean')
plt.fill_between(range(len(test_data)),
                 detector.baseline_mean - detector.threshold*detector.baseline_std,
                 detector.baseline_mean + detector.threshold*detector.baseline_std,
                 color='gray', alpha=0.2, label='Normal Range')
plt.ylabel('Mean Value')
plt.legend()
plt.grid(True)

# Plot the anomaly flags
plt.subplot(3, 1, 3)
anomaly_indices = np.where(anomaly_flags_padded)[0]
plt.plot(anomaly_indices, [1] * len(anomaly_indices), 'ro', label='Detected Anomaly', markersize=4)
plt.title('Anomaly Detection Results')
plt.xlabel('Sample Index')
plt.yticks([0, 1], ['Normal', 'Anomaly'])
plt.ylim(-0.5, 1.5)
plt.legend(loc='upper right')
plt.grid(True)

plt.tight_layout()
plt.show()

# @title Save Model Parameters
model_params = {
    "baseline_mean": float(detector.baseline_mean),
    "baseline_std": float(detector.baseline_std),
    "threshold": detector.threshold,
    "window_size": window_size
}

# Save to a file in Colab environment
with open('model_parameters.json', 'w') as f:
    json.dump(model_params, f, indent=4)

print("Model parameters saved to 'model_parameters.json'")

# @title Download Files from Colab
from google.colab import files

# Download the model parameters
files.download('model_parameters.json')

print("You can now download the model parameters file.")
```
